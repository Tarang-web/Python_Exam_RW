```python

import seaborn as sns
import matplotlib.pyplot as plt

# Set theme for consistency
sns.set_theme(style="darkgrid")

# 1 Count Plot - Passenger Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=titanic_df, palette='viridis')
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")
plt.show()

# 2 Histogram - Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 3 Box Plot - Fare by Class
plt.figure(figsize=(8, 5))
sns.boxplot(x="Pclass", y="Fare", data=titanic_df, palette="coolwarm")
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()

# 4 Bar Chart - Survival Count by Gender
plt.figure(figsize=(8, 5))
sns.barplot(x="Sex", y="Survived", data=titanic_df, ci=None, palette="muted")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

# 5 Violin Plot - Age Distribution by Class
plt.figure(figsize=(8, 5))
sns.violinplot(x="Pclass", y="Age", data=titanic_df, palette="pastel")
plt.title("Age Distribution Across Passenger Classes")
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()

# 6 Pie Chart - Survival Distribution
plt.figure(figsize=(6, 6))
titanic_df['Survived'].value_counts().plot.pie(autopct="%1.1f%%", colors=["red", "green"])
plt.title("Survival Distribution")
plt.ylabel("")
plt.show()

# 7 Scatter Plot - Age vs. Fare (Colored by Survival)
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=titanic_df, palette="coolwarm")
plt.title("Age vs. Fare (Survival Highlighted)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python


# Ensure correct column names (case-sensitive)
g = sns.catplot(x='Pclass', data=titanic_df, hue='Sex', kind='count', height=6, aspect=1.5)

# Set axis labels
g.set_axis_labels("Passenger Class", "Count")

# Set correct class labels (optional)
g.set_xticklabels(["Class 1", "Class 2", "Class 3"])

# Add title using `g.figure`
g.figure.suptitle("Passenger Class Distribution by Sex")

# Show the plot
plt.show()
```
