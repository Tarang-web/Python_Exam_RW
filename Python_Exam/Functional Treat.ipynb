```python
def S_M():  # show menu
    #  function main menu dikhata hai
    print("\nWelcome to the Data Analyzer Program!")
    print("1. Enter Data")
    print("2. Show Data Summary")
    print("3. Calculate Factorial")
    print("4. Filter Data")
    print("5. Sort Data")
    print("6. Show Data Stats")
    print("7. Exit")


def E_D():  # enter data
    # User se data input lene aur store karne ka kaam karta hai
    global d  # 'd' = 'data', jisme user ka input store hota hai
    d = list(map(int, input("Enter numbers separated by spaces: ").split()))
    print("\nData saved successfully!")


def S_S():  # show summary
    # Data ka summary dikhata hai jaise count, min, max, sum aur avg
    print("\nData Summary:")
    print(f"- Total items: {len(d)}")
    print(f"- Min: {min(d)}")
    print(f"- Max: {max(d)}")
    print(f"- Sum: {sum(d)}")
    print(f"- Avg: {sum(d) / len(d):.2f}")


def F(n):  # factorial
    # Ye recursive function hai jo factorial nikalta hai
    if n <= 1:
        return 1
    return n * F(n - 1)


def C_F():  # calculate factorial
    # User se ek number le ke uska factorial calculate karta hai
    num = int(input("\nEnter a number for factorial: "))  # 'num' user se input number leta hai
    print(f"Factorial of {num} is: {F(num)}")


def F_D():  # filter data
    # User se threshold input le kar data ko filter karta hai
    t = int(input("\nEnter a threshold: "))  # 't'='threshold', jo filter ke liye hai
    f_d = list(filter(lambda x: x >= t, d))  # 'f_d' filtered data ko store karta hai
    print(f"Filtered Data: {f_d}")


def S_D():  # sort data
    # Data ko ascending ya descending order mein sort karta hai
    print("\n1. Ascending\n2. Descending")
    ch = int(input("Choose sorting order: "))  # 'ch' = 'choice', user se sorting order ka input krtahai
    s_d = sorted(d, reverse=(ch == 2))  # 's_d' sorted data ko store karta hai
    print(f"Sorted Data: {s_d}")


def S_T():  # show stats
    # Data ke stats dikhata hai: min, max, sum aur avg
    print("\nData Stats:")
    print(f"- Min: {min(d)}")
    print(f"- Max: {max(d)}")
    print(f"- Sum: {sum(d)}")
    print(f"- Avg: {sum(d) / len(d):.2f}")


def M():  # main function
    # Main program loop hai jisme menu dikhata hai aur user se input leta hai
    global d  # 'd' ='data', jo globally store hota hai
    d = []

    while True:
        S_M()  # Menu dikhane ke liye function call
        ch = input("\nEnter your choice: ")  # 'ch' = 'choice', user se menu option ka input

        # User ki choice ke according action lena
        match ch:
            case "1":
                E_D()  # Data enter karne ke liye
            case "2":
                S_S() if d else print("No data. Please enter data first.")  # Agar data nahi hai to enter karne ka kehna
            case "3":
                C_F()  # Factorial calculate karne ka option
            case "4":
                F_D() if d else print("No data. Please enter data first.")  # Agar data nahi hai to filter karne ka kehna
            case "5":
                S_D() if d else print("No data. Please enter data first.")  # Agar data nahi hai to sort karne ka kehna
            case "6":
                S_T() if d else print("No data. Please enter data first.")  # Agar data nahi hai to stats dikhane ka kehna
            case "7":
                print("\n Thank you for using the data analyzer and transformer program . Goodbye!")  # Program ko band karne ka option
                break
            case _:
                print("Invalid choice. Try again.")  # Agar input galat hai to retry karne ka kehna


M()  #  main function ko call karta hai
```
