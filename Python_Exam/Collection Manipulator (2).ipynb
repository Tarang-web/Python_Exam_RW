```python
students = []
subjects_set = set()

print("Welcome to the Student Data Organizer!")

while True:
    print("\nSelect an option: ")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student Information")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("Enter student details: ")
        student_id = input("Student ID: ")
        name = input("Name: ")
        age = int(input("Age: "))
        grade = input("Grade: ")
        dob = input("Date of Birth (YYYY-MM-DD): ")
        subjects_input = input("Subjects (comma-separated): ").split("// 1 ")
        student_tuple = (student_id, dob)
        subjects_set.update(subjects_input)
        student_info = {
            'id': student_id,
            'name': name,
            'age': age,
            'grade': grade,
            'subjects': subjects_input
        }
        students.append(student_info)
        print("Student added successfully!")

    elif choice == "2":
        if not students:
            print("No students available.")
        else:
            print("\n--- Display All Students ---")
            for student in students:
                print(f"Student ID: {student['id']} | Name: {student['name']} | Age: {student['age']} | Grade: {student['grade']} | Subjects: {', '.join(student['subjects'])}")

    elif choice == "3":
        student_id = input("Enter student ID to update: ")
        student_found = False
        for student in students:
            if student['id'] == student_id:
                student_found = True
                print("Update details: ")
                student['age'] = int(input("New Age: "))
                student['subjects'] = input("New Subjects (comma-separated): ").split(", ")
                subjects_set.update(student['subjects'])
                print("Student information updated successfully!")
                break
        if not student_found:
            print("Student ID not found.")

    elif choice == "4":
        student_id = input("Enter student ID to delete: ")
        student_found = False
        for student in students:
            if student['id'] == student_id:
                students.remove(student)
                print(f"Student ID {student_id} deleted successfully!")
                student_found = True
                break
        if not student_found:
            print("Student ID not found.")

    elif choice == "5":
        print("\n--- Subjects Offered ---")
        if subjects_set:
            for subject in subjects_set:
                print(subject)
        else:
            print("No subjects available.")

    elif choice == "6":
        print("Thank you for using the Student Data Organizer!")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
students = []  # Ek empty list banaya hai jisme hum students ki details store karenge.
subjects_set = set()  # Ek empty set banaya hai jisme subjects ki list store hogi, bina duplicates ke.

print("Welcome to the Student Data Organizer!")  # User ko welcome message .

while True:  # Yeh ek infinite loop hai jo tab tak chalega jab tak user exit nahi kar leta.
    print("\nWhat would you like to do?")  # User se pucha hai ki wo kya action lena chahta hai.
    print("1. Add a New Student")  # Option 1: New student add karna
    print("2. View All Students")  # Option 2: Sab students ki details dekhna
    print("3. Modify Student Information")  # Option 3: Student ke details ko modify karna
    print("4. Remove a Student")  # Option 4: Student ko remove karna
    print("5. Show Available Subjects")  # Option 5: Available subjects ko dikhana
    print("6. Exit")  # Option 6: Program ko exit karna

    user_choice = input("Please enter the number of your choice: ")  # User se input liya

    match user_choice:  # Yeh match statement user ke choice ke basis par action decide karega.
        case "1":  # Agar user "1" input kare to
            student_id = input("Enter Student ID: ")
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))  # Student ki age input liya aur int me convert kiya
            grade = input("Enter Grade: ")
            dob = input("Enter Date of Birth (YYYY-MM-DD): ")
            subjects_input = input("Enter Subjects (separated by commas): ").split(", ")  # Subjects input liya hain aur split karke list banaya hai

            student = [  # Ek student ka record ek list me store kiya
                student_id, name, age, grade, dob, subjects_input  # Yeh list me student ki sari details hain.
            ]

            students.append(student)  # Yeh student ko students list me add kar  hai.
            subjects_set.update(subjects_input)  # Yeh subjects ko subjects_set me add kar raha hai, duplicates ko avoid karte

            print(f"Student {name} added successfully!")

        case "2":  # Agar user "2" input kare to
            if not students:  # Agar students list empty hai to
                print("There are no students to display.")
            else:
                print("\n--- Student List ---")
                for student in students:  # Sabhi students ko loop se print kiya  hai
                    subjects = student[5]  # Student ke subjects ko extract kiya  hai.
                    print(f"ID: {student[0]} | Name: {student[1]} | Age: {student[2]} | Grade: {student[3]} | Subjects: {', '.join(subjects)}")  # Student ke details display ho rahe hain.

        case "3":  # Agar user "3" input kare to
            student_id = input("Enter the ID of the student you want to update: ")  # Update karne ke liye student ID input kiya  hai.
            student_found = False

            for student in students:  # Students list me loop chalaya  hai.
                if student[0] == student_id:  # Agar student ka ID match karta hai
                    student_found = True  #  True set kiya jaa raha hai, matlab student mil gaya
                    student[2] = int(input("Enter New Age: "))  # Student ki age update  hai.
                    new_subjects = input("Enter New Subjects (separate by commas): ").split(", ")  # New subjects input liye  hai
                    student[5] = new_subjects  # New subjects ko student ke record me update kiya  hai

                    subjects_set.update(new_subjects)  # Subjects ko subjects_set me update kiya hai
                    print("Student details updated successfully!")  # Confirmation message dikhaya
                    break  # Loop se break kar rahe hain jab update kar diya.

            if not student_found:  # Agar student nahi milta to
                print("No student found with the given ID.")  # Message dikhaya raha hai ki student nahi mila

        case "4":  # Agar user "4" input kare to
            student_id = input("Enter the ID of the student to remove: ")  # Student ko remove karne ke liye ID input li  hai
            student_found = False  # Flag banaya gaya hai to check if student found

            for student in students:  # Loop chalaya hai students list me
                if student[0] == student_id:  # Agar student ka ID match karta hai
                    students.remove(student)  # Student ko list se remove kiya  hai
                    print(f"Student with ID {student_id} has been deleted.")
                    student_found = True  # Flag ko True set kar  hai
                    break  # Loop se break kar rahe hai

            if not student_found:  # Agar student nahi milta to
                print("Student with the given ID not found.")
        case "5":  # Agar user "5" input kare to
            print("\n--- Subjects Offered ---")  # Available subjects ka list dikhaya  hai
            if subjects_set:  # Agar koi subjects available hain
                for subject in subjects_set:  # Subjects list me loop chala rahe hain
                    print(subject)  # Har subject ko print kar rahe hain.
            else:
                print("No subjects available yet.")  # Agar koi subject nahi hai to message dikhaya  hai

        case "6":  # Agar user "6" input kare to
            print("Thank you for using the Student Data Organizer. Goodbye!")
            break

        case _:  # Agar koi invalid option di jaata hai
            print("Invalid choice! Please enter a valid option from the menu.")
```
