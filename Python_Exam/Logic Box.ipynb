```python
print("Welcome to the Pattern Generator and Number Analyzer!")

while True:
    print("\nSelect an option:")
    print("1. Generate a Pattern")
    print("2. Analyze a Range of Numbers")
    print("3. Exit")
    choice = input("Enter your choice : ")

    if choice == '1':
        print("\nChoose a pattern type:")
        print("1. Right-angled Triangle")
        print("2. Pyramid")
        print("3. Left-angled Triangle")
        pattern_choice = input("Enter your choice : ")

        if pattern_choice not in ['1', '2', '3']:
            print("Invalid pattern choice.")
        else:
            rows_input = input("Enter the number of rows: ")
            print("\n")

            if rows_input and int(rows_input) > 0:
                rows = int(rows_input)
                if pattern_choice == '1':
                    print("Pattern:")
                    for i in range(1, rows + 1):
                        for j in range(i):
                            print("*", end="")
                        print()

                elif pattern_choice == '2':
                    print("Pattern:")
                    for i in range(1, rows + 1):
                        for j in range(rows - i):
                            print(" ", end="")
                        for k in range(2 * i - 1):
                            print("*", end="")
                        print()

                elif pattern_choice == '3':
                    print("Pattern:")
                    for i in range(rows, 0, -1):
                        for j in range(i):
                            print("*", end="")
                        print()
            else:
                print("Error: Please enter a valid positive integer for rows.")

    elif choice == '2':
        print("\nNumber Analysis")
        start_input = input("Enter the start of the range: ")
        end_input = input("Enter the end of the range: ")


        if start_input and end_input.isnumeric():
            start = int(start_input)
            end = int(end_input)

            if start > end:
                print("Error: Please try again.")
            else:
                total_sum = 0
                for number in range(start, end + 1):
                    total_sum += number
                    if number % 2 == 0:
                        print(f"Number {number} is Even")
                    else:
                        print(f"Number {number} is Odd")

                print(f"Sum of all numbers from {start} to {end} is: {total_sum}")
        else:
            print("Error: Both start and end must be valid positive integers.")

    elif choice == '3':
        print("\nExiting the program. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
