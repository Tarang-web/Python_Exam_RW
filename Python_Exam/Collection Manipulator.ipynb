```python
students = []
subjects_set = set()

print("Welcome to the Student Data Organizer!")

while True:
    print("\nSelect an option: ")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student Information")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("Enter student details: ")
        student_id = input("Student ID: ")
        name = input("Name: ")
        age = int(input("Age: "))
        grade = input("Grade: ")
        dob = input("Date of Birth (YYYY-MM-DD): ")
        subjects_input = input("Subjects (comma-separated): ").split(", ")
        student_tuple = (student_id, dob)
        subjects_set.update(subjects_input)
        student_info = {
            'id': student_id,
            'name': name,
            'age': age,
            'grade': grade,
            'subjects': subjects_input
        }
        students.append(student_info)
        print("Student added successfully!")

    elif choice == "2":
        if not students:
            print("No students available.")
        else:
            print("\n--- Display All Students ---")
            for student in students:
                print(f"Student ID: {student['id']} | Name: {student['name']} | Age: {student['age']} | Grade: {student['grade']} | Subjects: {', '.join(student['subjects'])}")

    elif choice == "3":
        student_id = input("Enter student ID to update: ")
        student_found = False
        for student in students:
            if student['id'] == student_id:
                student_found = True
                print("Update details: ")
                student['age'] = int(input("New Age: "))
                student['subjects'] = input("New Subjects (comma-separated): ").split(", ")
                subjects_set.update(student['subjects'])
                print("Student information updated successfully!")
                break
        if not student_found:
            print("Student ID not found.")

    elif choice == "4":
        student_id = input("Enter student ID to delete: ")
        student_found = False
        for student in students:
            if student['id'] == student_id:
                students.remove(student)
                print(f"Student ID {student_id} deleted successfully!")
                student_found = True
                break
        if not student_found:
            print("Student ID not found.")

    elif choice == "5":
        print("\n--- Subjects Offered ---")
        if subjects_set:
            for subject in subjects_set:
                print(subject)
        else:
            print("No subjects available.")

    elif choice == "6":
        print("Thank you for using the Student Data Organizer!")
        break

    else:
        print("Invalid choice. Please try again.")
```
